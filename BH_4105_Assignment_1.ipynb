```python
# Code for ECGR 4105 Assignment 1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read data file and create variables containing required data
data = pd.read_csv('D3.csv', na_values='*')
X1 = data['X1'].values
X2 = data['X2'].values
X3 = data['X3'].values
Y = data['Y'].values

def gradient_descent(x, y, learning_rate, num_iterations):
  m = len(y)
  theta = 0


  loss_history = []
  theta_history = []

  for iteration in range(num_iterations):
    predictions = x * theta

    loss = np.mean((predictions - y) ** 2)
    loss_history.append(loss)

    gradient = (2/m) * np.dot(x.T, (predictions -y))
    theta = theta - learning_rate * gradient
    theta_history.append(theta)

    return theta, loss_history, theta_history


num_iterations = 1500
learning_rates = [0.1, 0.05, 0.01]

# Problem 1
for X, variable_name in zip([X1, X2, X3], ['X1', 'X2', 'X3']):
    plt.figure(figsize=(12, 6))

    # Plot data points
    plt.subplot(1, 2, 1)
    plt.scatter(X, Y, label='Data points')
    plt.xlabel(variable_name)
    plt.ylabel('Y')

    for learning_rate in learning_rates:

        num_iterations = 1000

        # Perform gradient descent
        theta, loss_history, theta_history = gradient_descent(X, Y, learning_rate, num_iterations)

        # Plot final regression model
        plt.plot(X, theta * X, label=f'Final Model (LR: {learning_rate})')

        # Plot loss over iterations
        plt.subplot(1, 2, 2)
        plt.plot(loss_history, label=f'Learning Rate: {learning_rate}')

    plt.xlabel('Iteration')
    plt.ylabel('Loss')
    plt.legend()
    plt.title(f'Final Regression Models and Loss for {variable_name}')
    plt.tight_layout()
    plt.show()

# Problem 2
combined_X = np.column_stack((X1, X2, X3))
combined_Y = np.column_stack((Y, Y, Y))
plt.figure(figsize=(12, 6))

    # Plot data points
plt.subplot(1, 2, 1)
plt.scatter(combined_X, combined_Y, label='Data points')
plt.xlabel('X')
plt.ylabel('Y')

learning_rates = 0.05

num_iterations = 1000

      # Perform gradient descent
theta, loss_history, theta_history = gradient_descent(combined_X, combined_Y, learning_rates, num_iterations)

      # Plot final regression model
plt.plot(combined_X[:, 0], np.dot(combined_X, theta), label=f'Final Model (LR: {learning_rates})')

      # Plot loss over iterations
plt.subplot(1, 2, 2)
plt.plot(loss_history, label=f'Learning Rate: {learning_rates}')

plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.legend()
plt.title(f'Final Regression Models and Loss for combined_X')
plt.tight_layout()
plt.show()







```
